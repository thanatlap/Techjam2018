```python
import numpy 
import pandas
```

```python
#credit_sum = pandas.read_csv('../prep_data/credit_summary.csv', index_col='ip_id')
#credit_sum_test = pandas.read_csv('../prep_data/credit_summary_test.csv', index_col='ip_id')
sa_sum1 = pandas.read_csv('../prep_data/SA_summary.csv', index_col='ip_id')
sa_sum2 = pandas.read_csv('../prep_data/summary_sa.csv', index_col='ip_id')
sa_sorn_train = pandas.read_csv('../prep_data/sa_summary_sorn_train.csv', index_col='ip_id')
sa_sorn_test = pandas.read_csv('../prep_data/sa_summary_sorn_test.csv', index_col='ip_id')

df_y_train = pandas.read_csv('../data/y_train.csv',index_col='ip_id')
df_y_test = pandas.read_csv('../data/y_test_index.csv',index_col='ip_id')
```

```python
alldf = pandas.concat([ sa_sum1, sa_sum2, sa_sorn_train, sa_sorn_test], axis=1)
```

```python
alldf
```

```python
all_df = alldf.drop('label', axis=1)
all_df['is_train'] = numpy.where(all_df.index.isin(df_y_train.index),1,0)
```

```python
num_df = all_df[all_df['is_train']==0].fillna(0).select_dtypes(['number']) 
```

```python
X = num_df.values
x_test = num_df.values
```

```python
from keras.layers import Input, Dense
from keras.models import Model
from keras import regularizers
input_img = Input(shape=(85,))
encoded = Dense(128, activation='relu')(input_img)
encoded = Dense(64, activation='relu')(encoded)
encoded = Dense(32, activation='relu', name='encoder')(encoded)

decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(128, activation='relu')(decoded)
decoded = Dense(85, activation='sigmoid')(decoded)
```

```python
autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
encoder = Model(input_img, encoded)
autoencoder.fit(X, X,
                epochs=100,
                batch_size=256)
```

```python
num_df_train = all_df[all_df['is_train']==1].fillna(0).select_dtypes(['number']) 
X = num_df_train.values
Y = df_y_train.values
```

```python
#encoder.layers.pop()
#encoder.layers[-1].outbound_nodes = []
#encoder.outputs = [encoder.layers[-1].output]
output = encoder.get_layer('encoder').output
output = Dense(output_dim=1, activation='sigmoid', name='output',kernel_regularizer=regularizers.l2(0.01),
                activity_regularizer=regularizers.l1(0.01))(output) # your newlayer Dense(...)
new_model = Model(encoder.input, output)
```

```python
from sklearn.utils import class_weight
sample_weights = class_weight.compute_sample_weight('balanced', Y)
```

```python
new_model.compile(optimizer='nadam', loss='binary_crossentropy', metrics=['accuracy'])
for layer in new_model.layers:
    if layer.name != 'output':
        layer.trainable = False
        
```

```python
new_model.fit(X, Y,
                epochs=100,
                batch_size=256,
             sample_weight=sample_weights)
```
